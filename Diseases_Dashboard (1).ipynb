```python
import pandas as pd
import numpy as np
import panel as pn
pn.extension('tabulator')
import hvplot.pandas
import holoviews as hv
from holoviews import opts

```

```python
df= pd.read_csv('C:/Users/skc/Downloads/Behavioral_Health_Q1_Q2_2021_THCIC.csv')
```

```python
df
```

```python
df= df.fillna(0)
df
```

```python
# Select relevant columns
df_diseases = df[['RACE', 'APR_DRG_CodeDesc']]
```

```python
df.columns
```

```python
df_diseases.head()
```

```python
# Define the diseases of interest
diseases_of_interest = ['Bipolar Disorders', 'Schizophrenia','MAJOR DEPRESSIVE DISORDERS AND OTHER OR UNSPECIFIED PSYCHOSES']
# Filter the DataFrame
df_filtered = df_diseases[df_diseases['APR_DRG_CodeDesc'].isin(diseases_of_interest)]
```

```python
df_filtered
```

```python
# Aggregate data: count occurrences of each APR_DRG_CodeDesc per RACE
df_aggregated = df_filtered.groupby(['RACE','APR_DRG_CodeDesc']).size().reset_index(name='Count')
```

```python
plot = df_aggregated.hvplot.bar(
    x='RACE',
    y='Count',
    by='APR_DRG_CodeDesc',
    title="Count of Bipolar Disorder, Major Depressive Disorder, and Schizophrenia by Race",
    xlabel='Race',
    ylabel='Count of Diseases',
    width=1000,
    height=400,
    tools=['hover'],
    stacked=False
)

# Display the plot
plot
```

```python
scatter_plot = df_aggregated.hvplot.scatter(
    x='RACE',
    y='Count',
    color='APR_DRG_CodeDesc',
    title="Disease Counts By Race",
    xlabel='Race',
    ylabel='Count of Diseases',
    width=800,
    height=400
)
```

```python
scatter_plot
```

```python
df_pivoted = df_aggregated.pivot(index='RACE', columns='APR_DRG_CodeDesc', values='Count').fillna(0)
```

```python
line_chart = df_pivoted.hvplot.line(
    title="Disease Counts by Race",
    xlabel='Race',
    ylabel='Count',
    width=1000,
    height=400,
    line_width=2,
    tools=['hover']
    
)
```
